# OMNIPOWER Linear Regression

### Importing all the libaries

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

### Contruct Dataframe

```python
df = pd.read_csv('OmniPower.csv')
df.head()
```

```python
x = df[['Price', 'Promotion']]  #must use double brackets to select df, becuz multiple col
y = df['Sales']

x_train = x
y_train = y

X_train, X_test, y_train, y_test = train_test_split(x_train,y_train, test_size = 0.2, random_state = 42)
print(X_train.head())
print(y_train.head())
```

```python
print(X_train.shape)
print(y_train.shape)
```

```python
model = LinearRegression()
model = model.fit(X_train, y_train)

print(model.intercept_)
print(model.coef_)
```

### Prediction

```python
#predicts on test data
y_pred = model.predict(X_train) #x_test and y_pred must be same length

#single value predict
print(model.predict([[50,200]]))
```

```python
y_pred
```

### Validate

```python
print(model.score(X_train, y_train))
```

### R2 score

```python
from sklearn.metrics import r2_score
r2 = r2_score(y_train, y_pred)
print("Test R2_score:", r2)
```
